```python
import numpy as np
import pandas as pd
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
u = (-20, 20)
pi = (.3, .7)
sig = (10, 10)

n1 = norm(loc=u[0], scale=sig[0])
n2 = norm(loc=u[1], scale=sig[1])

def pdf_mix(x):
    return n1.pdf(x)*pi[0] + n2.pdf(x)*pi[1] 

xs = np.linspace(-100,100,500)
x_pdf = pdf_mix(xs)
```

```python
fig, ax = plt.subplots()
ax.plot(xs,x_pdf,color='red',alpha=.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(linestyle='--',color='grey')
ax.set_yticks([0,0.01,0.02,0.03])
ax.set_title("The 'intractable' $P_M$")
fig.savefig('Pm2humps.png')
pass
```
